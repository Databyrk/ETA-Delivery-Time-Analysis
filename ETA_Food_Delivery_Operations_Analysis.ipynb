```python
def haversine_distance(lat1, lon1, lat2, lon2):
    # Radius of Earth in kilometers
    R = 6371  
    
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    return R * c
```

```python
import numpy as np

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    return R * c

df['distance_km'] = haversine_distance(
    df['Restaurant_latitude'],
    df['Restaurant_longitude'],
    df['Delivery_location_latitude'],
    df['Delivery_location_longitude']
)
```

```python
df[['distance_km']].head()
```


### Average Delivery Time by Distance

```python
df['distance_bucket'] = pd.cut(
    df['distance_km'],
    bins=[0, 2, 5, 10, 20, 50],
    labels=['0-2 km', '2-5 km', '5-10 km', '10-20 km', '20+ km']
)
```

```python
distance_kpi = df.groupby('distance_bucket')['delivery_time'].mean().reset_index()
distance_kpi
```


### Average delivery time by Type_of_order

```python
restaurant_kpi = (
    df.groupby('Type_of_order')['delivery_time']
      .mean()
      .sort_values(ascending=False)
      .reset_index()
)

restaurant_kpi


```

```python
restaurant_kpi = (
    df.groupby('Type_of_order')
      .agg(
          avg_delivery_time=('delivery_time', 'mean'),
          total_orders=('Type_of_order', 'count')
      )
      .sort_values('avg_delivery_time', ascending=False)
      .reset_index()
)

restaurant_kpi

```

```python
workload_balance_kpi = (
    df.groupby('Delivery_person_ID')
      .agg(
          total_orders=('Delivery_person_ID', 'count'),
          avg_delivery_time=('delivery_time', 'mean'),
          late_delivery_rate=('late_delivery', 'mean')
      )
      .reset_index()
)

workload_balance_kpi.head()
```

```python

workload_balance_kpi['late_delivery_rate'] *= 100
```

```python
balance_summary_kpi = {
    'avg_orders_per_person': df.groupby('Delivery_person_ID').size().mean(),
    'std_orders_per_person': df.groupby('Delivery_person_ID').size().std(),
    'avg_delivery_time': df['delivery_time'].mean(),
    'std_delivery_time': df['delivery_time'].std()
}

balance_summary_kpi
```

## 🔍 Key Insights

- Average delivery time increases significantly with distance.
- Certain order types consistently take longer to deliver.
- Late deliveries rise sharply beyond mid-range distance buckets.
- Delivery partner workload varies, indicating potential imbalance.
- Standard deviation in orders per partner suggests scope for better allocation.


## 💡 Recommendations

- Reassign high-distance orders to experienced delivery partners.
- Introduce dynamic batching for short-distance orders.
- Optimize staffing during peak hours to reduce late deliveries.
- Use workload balance metrics to prevent delivery partner burnout.
